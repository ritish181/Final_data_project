```python
import pandas as pd

# Load the employee data
path = r'/home/ritish/projects/DATA_ENGINEERING/data-final-project/data_science\Updated_Merged_Data.csv'
final_data = pd.read_csv(path)

# Clean column names by removing leading/trailing spaces
final_data.columns = final_data.columns.str.strip()

# Define weights for each metric
weights = {
    'DiscussionCount': 0.2,
    'AverageRating': 0.25,
    'FeedbackCount': 0.2,
    'TotalCourseCount': 0.2,
    'AverageQuizScore': 0.25,
    'TotalTimeSpent': 0.1
}

# Calculate the Engagement Level
final_data['EngagementLevelScore'] = (
    final_data['DiscussionCount'] * weights['DiscussionCount'] +
    final_data['AverageRating'] * weights['AverageRating'] +
    final_data['FeedbackCount'] * weights['FeedbackCount'] +
    final_data['TotalCourseCount'] * weights['TotalCourseCount'] +
    final_data['AverageQuizScore'] * weights['AverageQuizScore'] +
    final_data['TotalTimeSpent'] * weights['TotalTimeSpent']
)

# Define thresholds for classification
high_threshold = 80  # Adjust this value as needed
medium_threshold = 50  # Adjust this value as needed

# Classify Engagement Level
final_data['EngagementLevel'] = pd.cut(
    final_data['EngagementLevelScore'],
    bins=[-float('inf'), medium_threshold, high_threshold, float('inf')],
    labels=['Low', 'Medium', 'High']
)

# Display the updated DataFrame with Engagement Level
print("Data with Engagement Level:")
print(final_data[['EmployeeID', 'Department', 'EngagementLevelScore', 'EngagementLevel']])

# Save the modified data back to CSV
output_path = r'/home/ritish/projects/DATA_ENGINEERING/data-final-project/data_science\Updated_Merged_Data_with_Engagement.csv'
final_data.to_csv(output_path, index=False)

print(f"Updated data saved to {output_path}")
```
